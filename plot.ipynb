```python
import matplotlib.pyplot as plt
import pandas as pd 
df = pd.read_csv('runs/detect/train13/results.csv')
data = dict()
for key in df.keys():
    data[key.replace(' ', '')] = df[key].to_numpy()
    
plt.figure(figsize=(6, 30))
Numplots = len(data.keys()) -1 
count = 1
for key in data.keys():
      if key != 'epoch' and 'lr' not in key:
            plt.subplot(Numplots, 3, count)
            plt.plot(data['epoch'], data[key])
            plt.xlabel('epoch')
            plt.ylabel(key)
            count += 1
plt.tight_layout()
plt.show()
```
